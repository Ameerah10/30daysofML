This notebook was used to explain the concepts of data visualization using matplotlib and working on some real life examples. This notebook is purely for educational purpose only.

```python
# import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating a dummy variable
x =[1,2,3]
y =[5,10,15]
```

```python
plt.plot(x,y, label='Amount Sold', color='yellow', marker='o', linewidth= 5, linestyle='dashed',
        markersize=15)

# plt.figure(figsize = (10,5))

plt.title('My First Ever Graph', fontdict ={'color': 'yellow',
                                           'fontsize': 20,
                                           'fontname':'DejaVu Sans'})



plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.xticks([0.5,1,1.5,2,2.5,3])
plt.yticks([2.5,5,7.5,10,12.5,15])


plt.legend(title = 'Legend', bbox_to_anchor = (1.05, 1))
plt.savefig('ONE-IH Graph.png')

plt.show()
```

```python
a = [2,5,6,8,10]
b = [5,10,15,20,25]

plt.plot(a,b, 'b*-.', label='Enjoyment')
plt.title('My Second Graph!', fontdict={'fontsize': 10,
                                       'color': 'red'})
plt.figure(figsize = (50,20))
plt.legend(title='Aye jije', bbox_to_anchor=(1,1))
plt.show()
```

```python
# creating a dummy variable
x1 = np.arange(0,4.5,0.5)
x1
```

```python
# creating a dummy variable
x2 = x1**2
x2
```

```python
plt.plot(x,y)
plt.plot(a,b)

plt.show()
```

```python
# Combinin two different datatypes on a plot
plt.plot(x1[:6],x2[:6], 'g*--')
plt.plot(x1[4:],x2[4:],'b^-')

plt.legend()
plt.show()
```

## Barchart with Matplotlib


```python
labels =['Jan','Feb','March']
values =[500, 1000, 1500]
```

```python
plt.figure(figsize = (5,3))
bars = plt.bar(labels, values, width=0.5, color='red')

plt.title('My beautiful Barchart!')

bars[0].set_hatch('*')

plt.xticks(rotation=90)

plt.show()
```
